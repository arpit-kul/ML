```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
import scipy.stats as st
```

```python
df1 = pd.read_csv('Train.csv')
from sklearn.preprocessing import LabelEncoder
#model.drop('Unnamed: 0',axis=1,inplace=True)

df1.head()
```

```python
df1['Policy End Date']=df1['Policy End Date'].astype('datetime64')
df1['Policy Start Date']=df1['Policy Start Date'].astype('datetime64')

df1['days']=(df1['Policy End Date']-df1['Policy Start Date']).dt.days

df1 = df1[['ID', 'Age','Gender', 'No_Pol', 'Subject_Car_Make', 'ProductName', 'days', 'Car_Category']]
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
iqr = st.iqr(df1['Age'])
q1 = df1['Age'].quantile(0.25)
q3 = df1['Age'].quantile(0.75)

ll = q1 - (3*(iqr))
ul = q3 + (3*iqr)
```

```python
df1 = df1[(df1['Age']>ll)&(df1['Age']<ul)]
```

```python
df1.shape
```

```python
gender = {'Male':'Male', 'Female':'Female', 'Entity':np.nan, 'Joint Gender':'others', 'NO GENDER':np.nan,'NOT STATED':np.nan, 'SEX':np.nan}

df1['Gender']=df1['Gender'].map(gender)
```

```python
df1.drop('ID',axis=1,inplace=True)
```

```python
df1.head()
```

```python
ll = LabelEncoder()
df1['ProductName'] = ll.fit_transform(df1['ProductName'])
```

```python
df1
```

```python
new = pd.get_dummies(df1['Gender'],drop_first=True)
```

```python
df1 = df1.join(new)
```

```python
df1.drop('Gender',axis=1,inplace = True)
```

```python
df1
```

```python
df1 = df1.reset_index(drop=True)
```

```python
df3
```

```python
df3 = df3.reset_index(drop=True)
```

```python
final_train = pd.concat([df1,df3],axis=0)
```

```python
final_train
```

```python
final_train.reset_index(drop=True,inplace=True)
```

```python
l2_sub_car = LabelEncoder()
final_train['Subject_Car_Make'] = l2_sub_car.fit_transform(final_train['Subject_Car_Make'])
```

```python
final_train.dropna(axis=0,inplace=True)
```
